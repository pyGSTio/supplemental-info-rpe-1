```python
#Import relevant name spaces
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#Define auxiliary functions, taken from PRA 92, 062315.

#Equation 5.16 from PRA 92, 062315; we have relabeled M_j as N to be consistent with our notation.
def pmax_516(N,delta):
    return np.min([1,
           (1-.5*(1-np.sqrt(8)*delta)**2)**N / 
           (np.sqrt(2 * np.pi)*(1-np.sqrt(8)*delta)*np.sqrt(N))])

#Equations 5.7 and 5.8 from PRA 92, 062315.
def xi(h):
    return 2*np.pi/2**h

def xi_bar(K):
    return 2*np.pi/2**(K+1)

#Equation 5.9 from PRA 92, 062315.
def RMSE(K,N,delta):
    return np.sqrt(
        (1-pmax_516(N,delta))*xi_bar(K)**2+np.sum(xi(j)*pmax_516(N,delta) for j in xrange(1,K+1)))
```

```python
#Plot maximum RMS error as function of additive error delta
#for a range of L_max and N values.

#K here denotes log_2(L_max)+1
delta_range = np.arange(0,1/np.sqrt(8),.001)

K=5
N=4
plt.plot(delta_range,[RMSE(K,N,delta) for delta in delta_range],label='K='+str(K)+',N='+str(N)+'')
K=11
N=4
plt.plot(delta_range,[RMSE(K,N,delta) for delta in delta_range],label='K='+str(K)+',N='+str(N)+'')

K=5
N=16
plt.plot(delta_range,[RMSE(K,N,delta) for delta in delta_range],label='K='+str(K)+',N='+str(N)+'')
K=11
N=16
plt.plot(delta_range,[RMSE(K,N,delta) for delta in delta_range],label='K='+str(K)+',N='+str(N)+'')

K=5
N=370
plt.plot(delta_range,[RMSE(K,N,delta) for delta in delta_range],label='K='+str(K)+',N='+str(N)+'')
K=11
N=370
plt.plot(delta_range,[RMSE(K,N,delta) for delta in delta_range],label='K='+str(K)+',N='+str(N)+'')

plt.legend(loc='best')
plt.xlabel(r'$\delta$')
plt.ylabel(r'RMS error upper bound')
```
